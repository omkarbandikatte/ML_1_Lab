```python
class LogisticRegression:
  def sigmoid(self,z):
        sig = 1/(1+e**(-z))
        return sig
  def initialize(self,X):
        weights = np.zeros((shape(X)[1]+1,1))
        X = np.c_[np.ones((shape(X)[0],1)),X]
        return weights,X
  def fit(self,X,y,alpha=0.001,iter=400):
        weights,X = self.initialize(X)
        def cost(theta):
            z = dot(X,theta)
            cost0 = y.T.dot(log(self.sigmoid(z)))
            cost1 = (1-y).T.dot(log(1-self.sigmoid(z)))
            cost = -((cost1 + cost0))/len(y)
            return cost
        cost_list = np.zeros(iter,)
        for i in range(iter):
            weights = weights - alpha*dot(X.T,self.sigmoid(dot(X,weights))-np.reshape(y,(len(y),1)))
            cost_list[i] = cost(weights)
        self.weights = weights
        return cost_list
  def predict(self,X):
        z = dot(self.initialize(X)[1],self.weights)
        lis = []
        for i in self.sigmoid(z):

            if i>0.5:
                lis.append(1)
            else:
                lis.append(0)
        return lis
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression as SklearnLogReg
from sklearn.metrics import f1_score as sklearn_f1_score

# Load the dataset
new_df = pd.read_csv(r"/content/Iris(in).csv")
new_df = new_df[new_df['Species'] != 'Iris-setosa']

# Features and target variables
X = new_df.drop('Species', axis=1)
y = new_df['Species']

# Encode target variable 'Species' to numeric values (0 and 1 for binary classification)
y = y.map({'Iris-versicolor': 1, 'Iris-virginica': 0})

# Split the dataset
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.1, random_state=42)

# Standardizing the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_tr = scaler.fit_transform(X_tr)
X_te = scaler.transform(X_te)

# Custom Logistic Regression Implementation
class LogisticRegression:
    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))  # Corrected from 'e' to np.exp

    def initialize(self, X):
        weights = np.zeros((X.shape[1] + 1, 1))  # Initialized weights to zero
        X = np.c_[np.ones((X.shape[0], 1)), X]  # Add a column of ones for the intercept
        return weights, X

    def fit(self, X, y, alpha=0.001, iterations=400):
        weights, X = self.initialize(X)

        def cost(theta):
            z = np.dot(X, theta)
            cost_value = - np.mean(y * np.log(self.sigmoid(z)) + (1 - y) * np.log(1 - self.sigmoid(z)))
            return cost_value

        for i in range(iterations):
            z = np.dot(X, weights)
            weights -= alpha * np.dot(X.T, self.sigmoid(z) - y.reshape(-1, 1)) / len(y)

        self.weights = weights
        return self

    def predict(self, X):
        X = np.c_[np.ones((X.shape[0], 1)), X]  # Add intercept term to X
        z = np.dot(X, self.weights)
        return [1 if i > 0.5 else 0 for i in self.sigmoid(z)]

# Train the custom logistic regression model
custom_log_reg = LogisticRegression()
custom_log_reg.fit(X_tr, y_tr)
y_train_pred = custom_log_reg.predict(X_tr)
y_test_pred = custom_log_reg.predict(X_te)

# Calculate F1 score (for custom model)
def f1_score_custom(y_true, y_pred):
    tp, tn, fp, fn = 0, 0, 0, 0
    for true, pred in zip(y_true, y_pred):
        if true == 1 and pred == 1:
            tp += 1
        elif true == 1 and pred == 0:
            fn += 1
        elif true == 0 and pred == 1:
            fp += 1
        elif true == 0 and pred == 0:
            tn += 1
    precision = tp / (tp + fp) if tp + fp != 0 else 0
    recall = tp / (tp + fn) if tp + fn != 0 else 0
    return 2 * precision * recall / (precision + recall) if precision + recall != 0 else 0

f1_train = f1_score_custom(y_tr, y_train_pred)
f1_test = f1_score_custom(y_te, y_test_pred)

print(f"Custom model F1 score on training data: {f1_train}")
print(f"Custom model F1 score on test data: {f1_test}")

# Now using sklearn's Logistic Regression to compare the result
sklearn_model = SklearnLogReg()
sklearn_model.fit(X_tr, y_tr)
y_train_pred_sklearn = sklearn_model.predict(X_tr)
y_test_pred_sklearn = sklearn_model.predict(X_te)

f1_train_sklearn = sklearn_f1_score(y_tr, y_train_pred_sklearn)
f1_test_sklearn = sklearn_f1_score(y_te, y_test_pred_sklearn)

print(f"Sklearn Logistic Regression F1 score on training data: {f1_train_sklearn}")
print(f"Sklearn Logistic Regression F1 score on test data: {f1_test_sklearn}")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression as SklearnLogReg
from sklearn.metrics import f1_score as sklearn_f1_score

# Load the dataset
new_df = pd.read_csv(r"/content/Iris(in).csv")

# Filter out 'Iris-setosa' and focus on the binary classification task
new_df = new_df[new_df['Species'] != 'Iris-setosa']

# Features and target variables
X = new_df.drop('Species', axis=1)
y = new_df['Species']

# Encode the target variable 'Species' to numeric values (1 for 'Iris-versicolor', 0 for 'Iris-virginica')
y = y.map({'Iris-versicolor': 1, 'Iris-virginica': 0})

# Split the dataset into training and testing sets
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

# Standardizing the data
scaler = StandardScaler()

# Fit and transform the training data
X_tr = scaler.fit_transform(X_tr)

# Only transform the test data (do not fit again)
X_te = scaler.transform(X_te)
```

```python
class LogisticRegression:
    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))  # Corrected from 'e**(-z)' to np.exp(-z)

    def initialize(self, X):
        # Adding intercept term (column of ones) to X and initializing weights
        weights = np.zeros((X.shape[1] + 1, 1))
        X = np.c_[np.ones((X.shape[0], 1)), X]  # Add a column of ones for the intercept
        return weights, X

    def fit(self, X, y, alpha=0.001, iterations=400):
        weights, X = self.initialize(X)

        def cost(theta):
            z = np.dot(X, theta)
            # Calculate cost using the log loss function
            cost_value = - np.mean(y * np.log(self.sigmoid(z)) + (1 - y) * np.log(1 - self.sigmoid(z)))
            return cost_value

        for i in range(iterations):
            z = np.dot(X, weights)
            weights -= alpha * np.dot(X.T, self.sigmoid(z) - y.reshape(-1, 1)) / len(y)

        self.weights = weights
        return self

    def predict(self, X):
        X = np.c_[np.ones((X.shape[0], 1)), X]  # Add intercept term to X
        z = np.dot(X, self.weights)
        return [1 if i > 0.5 else 0 for i in self.sigmoid(z)]
```
